if we like to create a class variable which will take data and give some sort of outcome as per need, how can we do that?

with the help of constructor-->helps class to take a data while creating an object

```python
class snehan1:
    def __init__(self , phone_number, email_id, student_id ): #constructor inbuilt method/function
    #       reference
    
        self.phone_number = phone_number
        self.email_id = email_id
        self.student_id = student_id
    
    def return_student_details(self):
        return self.phone_number , self.email_id , self.student_id
    
```

```python
rohan = snehan1(123456789,"abc@gmail.com",101)
```

```python
rohan.phone_number
```

```python
rohan.email_id
```

```python
rohan.student_id
```

```python
rohan.return_student_details()
```

```python
sohan = snehan1(23344455,"wer@gmail.com",123)
```

```python
sohan.phone_number
```

```python
sohan.return_student_details()
```

```python
class snehan2:
    def __init__(sahoo , phone_number, email_id, student_id ): #constructor inbuilt method/function
    #       reference
    
        sahoo.phone_number = phone_number
        sahoo.email_id = email_id
        sahoo.student_id = student_id
    
    def return_student_details(sahoo):
        return sahoo.phone_number , sahoo.email_id , sahoo.student_id
```

```python
mohan = snehan2(123456789,"mohan@gmail.com",234)
```

```python
mohan.phone_number
```

```python
mohan.return_student_details()
```

self is just a convention you can use any other word for the reference just like the above example ('sahoo' is used)

```python
class snehan3:
    def __init__(sahoo , phone_number, email_id, student_id ): #constructor inbuilt method/function
    #       reference
    
        sahoo.phone_number1 = phone_number
        sahoo.email_id = email_id
        sahoo.student_id = student_id
    
    def return_student_details(sahoo):
        return sahoo.phone_number1 , sahoo.email_id , sahoo.student_id
```

```python
rohan = snehan3(123456,"rohan@gmail.com",123)
```

```python
rohan.phone_number
```

this means that class only understand only that side of the variable which is associated with self(pointer variable)

the other side is just an assignment

```python
rohan.return_student_details()
```
